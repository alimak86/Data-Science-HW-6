```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
from sklearn.metrics import pairwise_distances
from sklearn.decomposition import PCA
```

```python
data2D = pd.read_csv("data/data_2d.csv")
mnist = pd.read_csv("data/mnist.csv")
pca = PCA(2)
data2D = pca.fit_transform(data2D)
mnist = pca.fit_transform(mnist)
```

```python
max_clusters = 100 ##maximum expected clusters number
```

```python
def define_optimal_clustering(data, title):
    Sum_of_squared_distances = []
    K = range(1,max_clusters)
    for num_clusters in K :
     kmeans = KMeans(n_clusters=num_clusters)
     kmeans.fit(data)
     Sum_of_squared_distances.append(kmeans.inertia_)
    plt.plot(K,Sum_of_squared_distances,"bx-")
    plt.xlabel('Values of K, clusters') 
    plt.ylabel('Sum of squared distances/Inertia') 
    plt.title(title)
    return plt
```

```python
plot_data2D = define_optimal_clustering(data2D, "Elbow Method For Optimal k (data2D)")
plot_data2D.show()
```

```python
plot_mnist = define_optimal_clustering(mnist, "Elbow Method For Optimal k (mnist)")
plot_mnist.show()
```

```python
k = 100
```

```python
kmeans = KMeans(n_clusters = k)
```

```python
data_labels = kmeans.fit_predict(data2D)
```

```python
mnist_labels = kmeans.fit_predict(mnist)
```

```python
plt.scatter(data2D[:,0],data2D[:,1],c = data_labels)
## optimal clusters for data2D
```
